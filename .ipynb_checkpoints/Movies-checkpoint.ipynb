### Relationship between Rotten Tomatoes and IMDb ratings
1. Converted both columns into numeric
1. Removed all missing values
1. Called the histogram and trendline function to show distribution and relationship

```python
importlib.reload(mypytable)
importlib.reload(utils)
importlib.reload(plot_utils)

mv_file = os.path.join("input_data", "movies.csv")
mv_mypy = utils.load_data(mv_file)
mv_mypy.remove_rows_with_missing_values()
rt = mv_mypy.column_names.index("Rotten Tomatoes")
for i in range(len(mv_mypy.data)):
    rt_string = mv_mypy.data[i][rt].replace("%", "")
    rt_r = float(rt_string)
    mv_mypy.data[i][rt] = rt_r

plot_utils.histo(mv_mypy, "IMDb")
plot_utils.histo(mv_mypy, "Rotten Tomatoes")
plot_utils.trendline(mv_mypy, "Rotten Tomatoes", "IMDb")
```

There is a weak positive relationship with a correlation coef of 0.016 and cov of 17.8
